```python
import pandas as pd
import numpy as np

df = pd.read_csv('flights_all.csv')
df.head()
```

```python
print("VALORES NULOS:")
nulls = df.isnull().sum()
nulls_pct = (nulls / len(df) * 100).round(2)
null_df = pd.DataFrame({'Nulos': nulls, '% del total': nulls_pct})
print(null_df)
```

```python
# ============================================
# 1. INFORMACIÓN GENERAL
# ============================================
print(f"Filas:    {df.shape[0]:,}")
print(f"Columnas: {df.shape[1]}")
print(f"\nTipos de datos:")
print(df.dtypes)
print(f"\nEstadísticas descriptivas:")
df.describe()
```

```python
# ============================================
# 3. VUELOS POR AÑO
# ============================================
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

vuelos_año = df['YEAR'].value_counts().sort_index()
plt.figure(figsize=(10, 5))
sns.barplot(x=vuelos_año.index, y=vuelos_año.values, palette='Blues_d')
plt.title('Vuelos por Año (2015-2020)')
plt.xlabel('Año')
plt.ylabel('Número de Vuelos')
plt.tight_layout()
plt.show()
```

```python
# ============================================
# 4. VUELOS POR MES
# ============================================
vuelos_mes = df['MONTH'].value_counts().sort_index()
plt.figure(figsize=(10, 5))
sns.barplot(x=vuelos_mes.index, y=vuelos_mes.values, palette='Blues_d')
plt.title('Vuelos por Mes')
plt.xlabel('Mes')
plt.ylabel('Número de Vuelos')
plt.tight_layout()
plt.show()
```

```python
# ============================================
# 5. TOP 10 AEROLÍNEAS
# ============================================
top_airlines = df['OP_UNIQUE_CARRIER'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_airlines.values, y=top_airlines.index, palette='Blues_d')
plt.title('Top 10 Aerolíneas por Número de Vuelos')
plt.xlabel('Número de Vuelos')
plt.ylabel('Aerolínea')
plt.tight_layout()
plt.show()
```

```python
# ============================================
# 6. RETRASO PROMEDIO POR AEROLÍNEA
# ============================================
retraso_aerolinea = df.groupby('OP_UNIQUE_CARRIER')['ARR_DELAY_NEW'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=retraso_aerolinea.values, y=retraso_aerolinea.index, palette='Reds_d')
plt.title('Retraso Promedio en Llegada por Aerolínea (min)')
plt.xlabel('Minutos de retraso promedio')
plt.ylabel('Aerolínea')
plt.tight_layout()
plt.show()
```

```python
print(df['OP_UNIQUE_CARRIER'].unique())
```

```python
# ============================================
# 8. CANCELACIONES POR AÑO
# ============================================
cancelaciones = df.groupby('YEAR')['CANCELLED'].mean() * 100
plt.figure(figsize=(10, 5))
sns.lineplot(x=cancelaciones.index, y=cancelaciones.values, marker='o', color='red')
plt.title('% de Vuelos Cancelados por Año')
plt.xlabel('Año')
plt.ylabel('% Cancelados')
plt.tight_layout()
plt.show()
```

```python
# ============================================
# 7. ANÁLISIS DEL TARGET 2 — DELAYED_15
#    (Retraso en llegada >= 15 min)
# ============================================

# Crear la variable derivada solo para vuelos no cancelados
vuelos_operados = df[df['CANCELLED'] == 0].copy()
vuelos_operados['DELAYED_15'] = (vuelos_operados['ARR_DELAY_NEW'] >= 15).astype(int)

# --- Gráfico 1: % de vuelos con retraso >= 15 min por año ---
retraso_anual = vuelos_operados.groupby('YEAR')['DELAYED_15'].mean() * 100

plt.figure(figsize=(10, 5))
sns.lineplot(x=retraso_anual.index, y=retraso_anual.values, marker='o', color='steelblue')
plt.title('% de Vuelos con Retraso ≥ 15 min por Año')
plt.xlabel('Año')
plt.ylabel('% con Retraso Significativo')
plt.tight_layout()
plt.show()
```

```python
# --- Gráfico 2: % de vuelos con retraso >= 15 min por aerolínea ---
retraso_aerolinea = vuelos_operados.groupby('OP_UNIQUE_CARRIER')['DELAYED_15'].mean() * 100
retraso_aerolinea = retraso_aerolinea.sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=retraso_aerolinea.values, y=retraso_aerolinea.index, palette='Blues_d',
            hue=retraso_aerolinea.index, legend=False)
plt.title('% de Vuelos con Retraso ≥ 15 min por Aerolínea')
plt.xlabel('% con Retraso ≥ 15 min')
plt.ylabel('Aerolínea')
plt.tight_layout()
plt.show()
```

```python
# --- Resumen numérico ---
total_op = len(vuelos_operados)
total_delayed = vuelos_operados['DELAYED_15'].sum()
print(f"Vuelos operados (no cancelados): {total_op:,}")
print(f"Con retraso >= 15 min:           {total_delayed:,}  ({total_delayed/total_op*100:.2f}%)")
print(f"Sin retraso significativo:        {total_op - total_delayed:,}  ({(total_op-total_delayed)/total_op*100:.2f}%)")
```
